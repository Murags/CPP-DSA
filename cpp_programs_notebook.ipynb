## arrays.cpp

```c++
#include <iostream>
using namespace std;

int run_arrays() {
    int myArray[5];
    int average = 0;

    for (int i = 0; i < 5; i++) {
        cout << "Enter a value to populate the array: ";
        cin >> myArray[i];
        average += myArray[i];
    }

    average = average / 5;
    cout << "Average: " << average << endl;

    return 0;
}
```

Run with `run_arrays();`

## pseudocode.cpp

```c++
#include <stdio.h>

int run_pseudocode() {
    int x = 0;
    int y = 20;

    while (y > 6) {
        y = y - 4;
        x = x + 2;
    }

    printf("%d\n", x);

    return 0;
}
```

Run with `run_pseudocode();`

## upper_lower.cpp

```c++
#include <iostream>

int run_upper_lower() {
    std::cout << "Enter char: ";
    char c;
    std::cin >> c;

    int ascii = static_cast<int>(c);
    if (ascii >= 65 && ascii <= 90) {
        std::cout << "Uppercase" << std::endl;
    } else if (ascii >= 97 && ascii <= 122) {
        std::cout << "Lowercase" << std::endl;
    } else {
        std::cout << "Not a letter" << std::endl;
    }

    return 0;
}
```

Run with `run_upper_lower();`

## square.cpp

```c++
#include <iostream>

int perimeter(int l) {
    return 4 * l;
}

int area(int l) {
    return l * l;
}

int run_square() {
    int length = 0;
    std::cout << "Enter the length of the square: ";
    std::cin >> length;

    std::cout << "Perimeter: " << perimeter(length) << std::endl;
    std::cout << "Area: " << area(length) << std::endl;

    return 0;
}
```

Run with `run_square();`

## volume.cpp

```c++
#include <iostream>
#include <cmath>

int run_volume() {
    double raduis = 0;
    double volume = 0;

    std::cout << "Enter the radius: ";
    std::cin >> raduis;

    volume = 4.0 / 3.0 * 3.14159 * exp(raduis);
    std::cout << "Volume: " << volume << "\n";

    return 0;
}
```

Run with `run_volume();`

## days.cpp

```c++
#include <iostream>

int run_days() {
    int days = 0;

    std::cout << "Enter the number of days: ";
    std::cin >> days;

    std::cout << "\nTotal seconds: " << days * 24 * 60 * 60 << "\n";

    return 0;
}
```

Run with `run_days();`
